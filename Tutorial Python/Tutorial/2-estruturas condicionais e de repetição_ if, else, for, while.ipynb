No c√≥digo acima, a estrutura condicional `if-elif-else` √© utilizada para determinar a natureza das ra√≠zes da equa√ß√£o com base no valor de `delta`. O bloco de c√≥digo dentro do `if` √© executado quando `delta` √© maior que zero. ##  La√ßo `for`


O la√ßo `for` √© uma estrutura de controle em Python usada para iterar sobre uma sequ√™ncia de elementos, como listas, strings, tuplas, sets, entre outros. Ele permite executar um bloco de c√≥digo repetidamente para cada item na sequ√™ncia. O la√ßo `for` √© particularmente √∫til quando voc√™ precisa realizar uma a√ß√£o para cada elemento em uma cole√ß√£o de dados.

**Sintaxe B√°sica**

A sintaxe b√°sica do la√ßo for em Python √© a seguinte:



```python

    # Bloco de c√≥digo a ser executado
```
- `elemento` √© uma vari√°vel que receber√° cada item da sequ√™ncia em cada itera√ß√£o do la√ßo.

**Iterando sobre uma Lista**. Vamos come√ßar com um exemplo onde iremos iterar sobre uma lista de n√∫meros e realizar uma opera√ß√£o matem√°tica em cada elemento. Suponha que temos uma lista de n√∫meros e queremos calcular o quadrado de cada n√∫mero:

## A fun√ß√£o `enumerate()`
 fun√ß√£o `enumerate()` em Python √© usada para iterar sobre uma sequ√™ncia (como uma lista, tupla, ou string) ao mesmo tempo em que obt√©m o √≠ndice e o valor de cada elemento da sequ√™ncia. Ela retorna um objeto enumerado que cont√©m pares de √≠ndice-valor.

 **Sintaxe B√°sica**
```python

```
- O par√¢metro `iteravel` √© a sequ√™ncia que desejamos iterar e obter os √≠ndices e valores.

- O par√¢metro `start` √© opcional e define o valor inicial do √≠ndice. Por padr√£o, √© 0.

A fun√ß√£o `enumerate()` retorna um objeto enumerado que pode ser convertido em uma lista, tupla ou usado em um la√ßo for para iterar sobre os elementos da sequ√™ncia juntamente com seus √≠ndices. 





````python
nome = ['L', 'E', 'O', 'N']
list(enumerate(nome))
````

**Usando a fun√ß√£o `enumerate()`**. Dada a tabela a seguir, vamos calcular a m√©dia ponderada de uma lista de notas, em que cada nota possui um peso correspondente. 


| Notas | Pesos |
|-------|-------|
|  8.5  |  0.3  |
|  7.2  |  0.2  |
|  9.0  |  0.4  |
|  6.8  |  0.1  |

Podemos usar o la√ßo `for` em conjunto com a fun√ß√£o `enumerate()` para percorrer a lista de notas e seus respectivos pesos, calculando a soma dos produtos de cada nota pelo seu peso. Antes vamos exemplificar o que faz fun√ß√£o `enumerate()`.

````python
notas = [8.5, 7.2, 9.0, 6.8]
pesos = [0.3, 0.2, 0.4, 0.1]

soma_produtos = 0
soma_pesos = 0

for indice, nota in enumerate(notas):
    peso = pesos[indice]
    soma_produtos += nota * peso
    soma_pesos += peso

media_ponderada = soma_produtos / soma_pesos

print(f"A m√©dia ponderada das notas √©: {media_ponderada}")

````

## 3.4 La√ßo `while`

O la√ßo `while` √© usado para repetir um bloco de c√≥digo enquanto uma condi√ß√£o especificada for verdadeira. Isso permite que voc√™ execute um conjunto de instru√ß√µes v√°rias vezes, desde que a condi√ß√£o seja atendida. √â especialmente √∫til quando voc√™ n√£o sabe o n√∫mero exato de vezes que o c√≥digo precisa ser repetido.

**Sintaxe B√°sica**



```python
while condi√ß√£o:
    # bloco de c√≥digo a ser executado enquanto a condi√ß√£o for verdadeira
    # atualiza√ß√£o da condi√ß√£o (se necess√°rio)
```

**Aten√ß√£o!** üòß √â importante ter cuidado com a atualiza√ß√£o da condi√ß√£o dentro do bloco de c√≥digo, caso contr√°rio, o la√ßo pode se tornar infinito.

Aqui est√° um exemplo simples para ilustrar como o la√ßo while funciona:

````python
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1

print("Fim do programa")
````

No c√≥digo acima, temos uma vari√°vel `contador` inicializada com o valor zero. A condi√ß√£o especificada √© `contador < 5`, o que significa que o bloco de c√≥digo ser√° executado enquanto o valor do contador for menor que 5.

A cada itera√ß√£o, o valor do contador √© impresso e incrementado. O la√ßo continua at√© que a condi√ß√£o `contador < 5` seja falsa, ou seja, quando o contador atingir o valor 5.

Ap√≥s a conclus√£o do la√ßo, a mensagem "Fim do programa" √© exibida na tela.

## Instru√ß√µes `break` e `continue`

Neste tutorial, vamos explorar as instru√ß√µes `break` e `continue` em Python. Essas instru√ß√µes s√£o utilizadas para controlar o fluxo de execu√ß√£o em la√ßos de repeti√ß√£o, como `for` e `while`. Vamos entender como cada uma delas funciona e como aplic√°-las em situa√ß√µes espec√≠ficas.

### Instru√ß√£o `break`

A instru√ß√£o `break` √© utilizada para interromper a execu√ß√£o de um la√ßo de repeti√ß√£o de forma prematura. Quando a instru√ß√£o `break` √© encontrada dentro de um la√ßo, o programa sai imediatamente do la√ßo, ignorando o restante das itera√ß√µes. Isso pode ser √∫til quando queremos parar a execu√ß√£o de um la√ßo antes de percorrer todos os elementos ou quando uma condi√ß√£o espec√≠fica √© atendida.




**Interrompendo uma execu√ß√£o usando `break`**. Vamos cncontrar o primeiro n√∫mero divis√≠vel por 7 entre 100 e 200. 

````python
a = 100
b = 200

for numero in range(a, b):
    if numero % 7 == 0:
        print("O primeiro n√∫mero divis√≠vel por 7 √©:", numero)
        break
````

No c√≥digo acima, assim que encontramos o primeiro n√∫mero divis√≠vel por 7, usamos a instru√ß√£o `break` para interromper o la√ßo e exibimos o n√∫mero encontrado.

**Agora usando o `while`**. Vamos encontrar o menor quadrado perfeito maior que um n√∫mero dado.

````python
numero = int(input("informe um inteiro positivo =  "))
quadrado = 1

while True:
    if quadrado ** 2 > numero:
        print("O menor quadrado perfeito maior que", numero, "√©", quadrado ** 2)
        break
    quadrado += 1
````

No c√≥digo anterior, utilizamos um la√ßo `while` infinito para iterar sobre os poss√≠veis valores de quadrados. Dentro do la√ßo, verificamos se o quadrado atual √© maior do que o n√∫mero dado. Se for, encontramos o menor quadrado perfeito desejado e utilizamos a instru√ß√£o `break` para finalizar o la√ßo. Caso contr√°rio, incrementamos o valor do `quadrado` e continuamos o la√ßo at√© encontrar o resultado desejado. O uso adequado da instru√ß√£o `break` √© importante para evitar la√ßos infinitos e garantir a efici√™ncia do c√≥digo.

## Instru√ß√£o `continue`

A instru√ß√£o continue √© utilizada para pular a itera√ß√£o atual de um la√ßo de repeti√ß√£o e continuar para a pr√≥xima itera√ß√£o. Ao encontrar a instru√ß√£o continue, o programa volta ao in√≠cio do la√ßo e verifica a pr√≥xima itera√ß√£o, ignorando o restante do c√≥digo dentro do bloco do la√ßo para a itera√ß√£o atual.


**Ignorando itera√ß√µes com `continue`**. Vamos imprimir todos os n√∫meros impares entre 1 e 20.

````python
a = 1
b = 20

for numero in range(a, b):
    if numero % 2 == 0:
        continue
    print(numero)
````

No c√≥digo acima, a condi√ß√£o `numero % 2 != 0` verifica se o n√∫mero √© par. Se a condi√ß√£o for verdadeira, usamos a instru√ß√£o `continue` para pular para a pr√≥xima itera√ß√£o, ignorando os n√∫meros pares. Portanto, apenas os n√∫meros √≠mpares s√£o impressos.

## 3.5 Compreens√£o de lista ( La√ßos e filtros em listas)

Uma t√©cnica particularmente √∫til em **Python** √© a constru√ß√£o de listas usando compreens√µes de lista . Esse recurso √© muito parecido com a nota√ß√£o do construtor definido que usamos em matem√°tica. Com o **Python** podemos contruir parte desse conjunto usando uma sintaxe similar. A seguir contruimos a lista com inteiros pares de 0 at√© 20.

**Filtrando informa√ß√µes**. A compreens√£o de lista √© particularmente √∫til quando queremos filtrar informa√ß√µes. Agora vamos criar uma lista de pares que s√£o divis√≠veis por 2 e 3 e s√£o menores que 20.  O passo a seguir √© chamado de **filtro** de lista.

````python
[n for n in range(2, 21, 2) if n%3==0]
````

Obtendo a soma de todos eles usando a fun√ß√£o **_sum_**:

````python
sum([n for n in range(2, 21, 2) if n%3==0])
````

**Usando a fun√ß√£o `zip`**. Pareando listas obtidas por compreens√£o.Vamos obter uma lista contendo os pares $(x,\sin{x})$ para alguns valores  $0< x < \pi$. Aqui usamos a fun√ß√£o `round` para arredondar para o n√∫mero mais pr√≥ximo com 4 casas decimais.

````python
import math

A = [0, 1/4, 1/2, 2.0, round(math.pi,4) ]

B = [round(math.sin(i), 4) for i in A]

AxB = zip(A,B) # pareamento de listas
print(list(AxB))
````

A fun√ß√£o `zip` √© usada para combinar elementos de duas ou mais listas em pares ordenados. Ela cria um objeto iter√°vel que cont√©m tuplas formadas pelos elementos correspondentes das listas de entrada. Como vimos acima, se tivermos duas listas `A` e `B`, ao usar `zip(A, B)`, obteremos um iterador que retornar√° tuplas contendo o primeiro elemento de `A` e o primeiro elemento de `B`, o segundo elemento de `A` e o segundo elemento de `B`, e assim por diante.
