## Indentação

A indentação é o espaçamento no início de cada linha de código para indicar a qual bloco ela pertence. Em Python, a identação é obrigatória e é geralmente feita usando espaços. A quantidade de espaços utilizada deve ser consistente ao longo do código. É uma boa prática utilizar 4 espaços como padrão de indentação.

A indentação correta é fundamental para que o Python entenda a estrutura do código corretamente. As instruções com a mesma indentação são consideradas parte do mesmo bloco de código.
Neste tutorial, vamos apresentar a sintaxe da linguagem de programação python. Focaremos no essencial útil para matemática.

## Blocos de Código

Um bloco de código em Python é um conjunto de instruções que são executadas em conjunto. Os blocos de código permitem que você agrupe instruções relacionadas para que sejam executadas em sequência. Em Python, os blocos de código são delimitados pela indentação, que é o espaçamento no início de cada linha de código.

## Estrutura 

A estrutura básica de um programa em Python é composta por instruções que são organizadas em blocos de código. Os blocos de código são delimitados pela indentação, ou seja, pelo recuo dos espaços ou tabulações no início das linhas. A indentação é fundamental em Python para indicar a hierarquia e o agrupamento das instruções.

Aqui está um exemplo de como um programa em Python pode ser estruturado:



```python
if condição:
    # Instruções dentro do primeiro bloco de código (nível de indentação 1)
    instrução1
    instrução2

    if outra_condição:
        # Instruções dentro do segundo bloco de código (nível de indentação 2)
        instrução3
        instrução4

    # Mais instruções dentro do primeiro bloco de código
    instrução5

# Instruções fora do bloco de código condicional
instrução6
```
Observe que as instruções dentro de um bloco de código são executadas somente se a condição correspondente for verdadeira. A indentação adequada ajuda a identificar quais instruções estão aninhadas em cada bloco.

**Simples exemplo da estrutura**. A seguir um exemplo real de estrutura de um programa escrito na linguagem Python.


```python
if idade >= 18:
    print("Você é maior de idade.")
    print("Pode entrar.")
else:
    print("Você é menor de idade.")
    print("Acesso negado.")
print(idade)
```
Neste exemplo, as linhas de código que possuem a mesma indentação (4 espaços) fazem parte do mesmo bloco de código. No primeiro bloco, que está dentro do `if`, as linhas serão executadas apenas se a condição `idade >= 18 `for verdadeira. Já no segundo bloco, que está dentro do else, as linhas serão executadas apenas se a condição for falsa. Por fim, a linha `print(idade)` representa outro bloco de código que será executado independentemente das condições anteriores, pois não está indentada dentro de nenhuma estrutura condicional.

Essa estrutura de blocos de código ajuda a controlar o fluxo de execução do programa e a organizar as instruções de forma lógica e legível.

## A declaração `elif`.
A declaração `elif` que é uma abrevição do `if-else` e é utilizada para testar múltiplas condições em sequência, oferecendo uma alternativa para o bloco "if" e precedendo o bloco "else".

Reescrevendo o código anterior usando a abreviação `elif`.

````python
nota = 7.2
if nota >=7:
    print('Aprovado')
elif nota>=3:
    print('Recuperação')
else:
    print('Reprovado')
````

**OBS.** As condições são avaliadas em ordem e, uma vez que o primeiro é cumprido, o código associado é executado e encerrado em seguida.

**Classificação de um triângulo.** Vamos escrever um código que classifique um triângulo com base em suas medidas dos lados. 

````python
lado1 = 3
lado2 = 4
lado3 = 5

if lado1 == lado2 == lado3:
    print("O triângulo é equilátero.")
elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
    print("O triângulo é isósceles.")
else:
    print("O triângulo é escaleno.")
````

**Solução de equações de grau 2.** Vamos resolver uma equação quadrática e exibir suas raízes. Vamos fazer uso do módulo `math`para calcular a raiz quadrada.

````python
import math

a = 2
b = -7
c = 3

delta = b**2 - 4*a*c

if delta > 0:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("As raízes da equação são", x1, "e", x2)
elif delta == 0:
    x = -b / (2*a)
    print("A equação possui uma raiz dupla:", x)
else:
    print("A equação não possui raízes reais.")
````

No código acima, a estrutura condicional `if-elif-else` é utilizada para determinar a natureza das raízes da equação com base no valor de `delta`. O bloco de código dentro do `if` é executado quando `delta` é maior que zero. Neste caso, utiliza-se a fórmula de Bhaskara para calcular as raízes `x1` e `x2`, e em seguida exibe uma mensagem informando as raízes encontradas. Se `delta` for igual a zero, o bloco dentro do `elif` é executado. Utilizando novamente a fórmula de Bhaskara, determina-se o valor da única raiz `x` e imprime-se uma mensagem informando a raiz encontrada. Por fim, se `delta` for menor que zero, o bloco de código dentro do `else` é executado, e uma mensagem é exibida informando que a equação não possui raízes reais.

##  Laço `for`


O laço `for` é uma estrutura de controle em Python usada para iterar sobre uma sequência de elementos, como listas, strings, tuplas, sets, entre outros. Ele permite executar um bloco de código repetidamente para cada item na sequência. O laço `for` é particularmente útil quando você precisa realizar uma ação para cada elemento em uma coleção de dados.

**Sintaxe Básica**

A sintaxe básica do laço for em Python é a seguinte:



```python
for elemento in sequencia:
    # Bloco de código a ser executado
```
- `elemento` é uma variável que receberá cada item da sequência em cada iteração do laço.
-  `sequencia` é a sequência  de elementos (em geral, listas, tuplas, sets)sobre a qual o laço irá iterar.



**Iterando sobre uma Lista**. Vamos começar com um exemplo onde iremos iterar sobre uma lista de números e realizar uma operação matemática em cada elemento. Suponha que temos uma lista de números e queremos calcular o quadrado de cada número:

````python
numeros = [1, 2, 3, 4, 5]

for numero in numeros:
    quadrado = numero ** 2
    print(quadrado)
````

No código acima, o laço `for` irá iterar sobre a lista `numeros` e a variável `numero` receberá cada elemento da lista em cada iteração. O bloco de código dentro do `for` calcula o quadrado de cada número usando o operador de exponenciação `**` e imprime o resultado.

**Iterando sobre uma Sequência Numérica**. Além de iterar sobre listas e tuplas, o laço `for` também pode ser usado para iterar sobre uma sequência numérica usando a função `range()`. Vamos considerar um exemplo onde queremos imprimir os números de 1 a 5:

````python
for numero in range(1, 6):
    print(numero)
````

No código acima, a função `range(1, 6)` cria uma sequência numérica de `1` a `5` (o último número é exclusivo). O laço `for` itera sobre essa sequência e a variável `numero` recebe cada valor da sequência em cada iteração. O bloco de código dentro do `for` imprime cada número.

**Envolvendo `if`**. A seguir vamos imprimir números entre 12 e 100 que são divisiveis por 7:

````python
for n in range(13, 100):
    if n%7==0:                 # verifica se numero n deixa resto zero na divisão por 7.
        print(n)
````

No código acima, a função `range(13, 100)` cria uma sequência numérica de `13` a `99`. O laço `for` itera sobre essa sequência e a variável `n` recebe cada valor da sequência em cada iteração. No bloco de código dentro do `for` a instrução `if n%7==0  ` verifica se `n`é divisível por 7. Se a condição for verdadeira, o número `n` será exibido.

**Somando valores de uma Lista**. Vamos apresentar um código capaz de somar todos os valores de uma lista.

````python
numeros = [1, 2, 3, 4, 5]
soma = 0

for numero in numeros:
    soma += numero

print("A soma dos números é:", soma)
````

No código acima, temos uma lista de números `[1, 2, 3, 4, 5]` e queremos calcular a soma desses números. Inicialmente, definimos uma variável `soma` com o valor zero.

Em seguida, utilizamos um laço `for` para percorrer cada elemento da lista. A cada iteração, o número atual é adicionado à variável `soma` usando o operador de atribuição combinada `+=`. Isso atualiza o valor da variável `soma` somando o número atual a seu valor anterior. 

**Obs**. Para subtração, multiplicação ou divsão usamos os operadores `-=`, `*=`, `/=`, respectivamente.

Neste caso, resultado pode ser obtido usando a função `sum`. Veja a seguir:

````python
sum(numeros)
````

**Soma seletiva**. Agora vamos somar todos os divisíveis por 7 entre de 12 e 100.

````python
soma = 0
for numero in range(13, 100):
    if numero%7==0:
        soma += numero
print(soma)
````

## A função `enumerate()`
 função `enumerate()` em Python é usada para iterar sobre uma sequência (como uma lista, tupla, ou string) ao mesmo tempo em que obtém o índice e o valor de cada elemento da sequência. Ela retorna um objeto enumerado que contém pares de índice-valor.

 **Sintaxe Básica**
```python
enumerate(iteravel, start=0)

```
- O parâmetro `iteravel` é a sequência que desejamos iterar e obter os índices e valores.

- O parâmetro `start` é opcional e define o valor inicial do índice. Por padrão, é 0.

A função `enumerate()` retorna um objeto enumerado que pode ser convertido em uma lista, tupla ou usado em um laço for para iterar sobre os elementos da sequência juntamente com seus índices. 





````python
nome = ['L', 'E', 'O', 'N']
list(enumerate(nome))
````

**Usando a função `enumerate()`**. Dada a tabela a seguir, vamos calcular a média ponderada de uma lista de notas, em que cada nota possui um peso correspondente. 


| Notas | Pesos |
|-------|-------|
|  8.5  |  0.3  |
|  7.2  |  0.2  |
|  9.0  |  0.4  |
|  6.8  |  0.1  |

Podemos usar o laço `for` em conjunto com a função `enumerate()` para percorrer a lista de notas e seus respectivos pesos, calculando a soma dos produtos de cada nota pelo seu peso. Antes vamos exemplificar o que faz função `enumerate()`.

````python
notas = [8.5, 7.2, 9.0, 6.8]
pesos = [0.3, 0.2, 0.4, 0.1]

soma_produtos = 0
soma_pesos = 0

for indice, nota in enumerate(notas):
    peso = pesos[indice]
    soma_produtos += nota * peso
    soma_pesos += peso

media_ponderada = soma_produtos / soma_pesos

print(f"A média ponderada das notas é: {media_ponderada}")

````

## 3.4 Laço `while`

O laço `while` é usado para repetir um bloco de código enquanto uma condição especificada for verdadeira. Isso permite que você execute um conjunto de instruções várias vezes, desde que a condição seja atendida. É especialmente útil quando você não sabe o número exato de vezes que o código precisa ser repetido.

**Sintaxe Básica**



```python
while condição:
    # bloco de código a ser executado enquanto a condição for verdadeira
    # atualização da condição (se necessário)

```

**Atenção!** 😧 É importante ter cuidado com a atualização da condição dentro do bloco de código, caso contrário, o laço pode se tornar infinito.

Aqui está um exemplo simples para ilustrar como o laço while funciona:

````python
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1

print("Fim do programa")
````

No código acima, temos uma variável `contador` inicializada com o valor zero. A condição especificada é `contador < 5`, o que significa que o bloco de código será executado enquanto o valor do contador for menor que 5.

Dentro do bloco de código, exibimos o valor atual do contador usando a função `print()`. Em seguida, atualizamos o contador adicionando 1 ao seu valor atual usando o operador de atribuição combinada `+=`.

A cada iteração, o valor do contador é impresso e incrementado. O laço continua até que a condição `contador < 5` seja falsa, ou seja, quando o contador atingir o valor 5.

Após a conclusão do laço, a mensagem "Fim do programa" é exibida na tela.



**Usando `while` e `if`**. Vamos calcular a soma dos divisores de 7 entre 12 e 100.

````python
soma = 0
numero = 13
while numero<100:
    if numero%7==0:      
        soma += numero
    numero += 1
soma
````

Agora vamos contar quantos números entre 12 e 100 são divisíveis por 7.

````python
cont = 0
numero = 0
while numero <100:
    if numero%7==0:
        cont+=1
    numero+=1
print(cont)
````

**Usando a função `input`**. O laço `while` também pode ser usado para realizar cálculos matemáticos iterativos. vamos criar um programa que calcula o fatorial de um número usando o laço `while`:


````python
numero = int(input("Digite um número inteiro positivo: "))
fatorial = 1
contador = 1

while contador <= numero:
    fatorial *= contador
    contador += 1

print("O fatorial de", numero, "é", fatorial)
````

## Instruções `break` e `continue`

Neste tutorial, vamos explorar as instruções `break` e `continue` em Python. Essas instruções são utilizadas para controlar o fluxo de execução em laços de repetição, como `for` e `while`. Vamos entender como cada uma delas funciona e como aplicá-las em situações específicas.

### Instrução `break`

A instrução `break` é utilizada para interromper a execução de um laço de repetição de forma prematura. Quando a instrução `break` é encontrada dentro de um laço, o programa sai imediatamente do laço, ignorando o restante das iterações. Isso pode ser útil quando queremos parar a execução de um laço antes de percorrer todos os elementos ou quando uma condição específica é atendida.




**Interrompendo uma execução usando `break`**. Vamos cncontrar o primeiro número divisível por 7 entre 100 e 200. 

````python
a = 100
b = 200

for numero in range(a, b):
    if numero % 7 == 0:
        print("O primeiro número divisível por 7 é:", numero)
        break
````

No código acima, assim que encontramos o primeiro número divisível por 7, usamos a instrução `break` para interromper o laço e exibimos o número encontrado.

**Agora usando o `while`**. Vamos encontrar o menor quadrado perfeito maior que um número dado.

````python
numero = int(input("informe um inteiro positivo =  "))
quadrado = 1

while True:
    if quadrado ** 2 > numero:
        print("O menor quadrado perfeito maior que", numero, "é", quadrado ** 2)
        break
    quadrado += 1
````

No código anterior, utilizamos um laço `while` infinito para iterar sobre os possíveis valores de quadrados. Dentro do laço, verificamos se o quadrado atual é maior do que o número dado. Se for, encontramos o menor quadrado perfeito desejado e utilizamos a instrução `break` para finalizar o laço. Caso contrário, incrementamos o valor do `quadrado` e continuamos o laço até encontrar o resultado desejado. O uso adequado da instrução `break` é importante para evitar laços infinitos e garantir a eficiência do código.

## Instrução `continue`

A instrução continue é utilizada para pular a iteração atual de um laço de repetição e continuar para a próxima iteração. Ao encontrar a instrução continue, o programa volta ao início do laço e verifica a próxima iteração, ignorando o restante do código dentro do bloco do laço para a iteração atual.


**Ignorando iterações com `continue`**. Vamos imprimir todos os números impares entre 1 e 20.

````python
a = 1
b = 20

for numero in range(a, b):
    if numero % 2 == 0:
        continue
    print(numero)
````

No código acima, a condição `numero % 2 != 0` verifica se o número é par. Se a condição for verdadeira, usamos a instrução `continue` para pular para a próxima iteração, ignorando os números pares. Portanto, apenas os números ímpares são impressos.

## 3.5 Compreensão de lista ( Laços e filtros em listas)

Uma técnica particularmente útil em **Python** é a construção de listas usando compreensões de lista . Esse recurso é muito parecido com a notação do construtor definido que usamos em matemática. Por exemplo, o conjunto de inteiros pares pode ser escrito como: $$ \{2\cdot k \,\,|\,\,\, k \in \mathbb{Z}\}.$$
Com o **Python** podemos contruir parte desse conjunto usando uma sintaxe similar. A seguir contruimos a lista com inteiros pares de 0 até 20.

````python
[2*k for k in range(11)]
````

O resultado acima também poderia ser alcançado usando a função **_range_**:


````python
list(range(0,21, 2))
````

**Filtrando informações**. A compreensão de lista é particularmente útil quando queremos filtrar informações. Agora vamos criar uma lista de pares que são divisíveis por 2 e 3 e são menores que 20.  O passo a seguir é chamado de **filtro** de lista.

````python
[n for n in range(2, 21, 2) if n%3==0]
````

Obtendo a soma de todos eles usando a função **_sum_**:

````python
sum([n for n in range(2, 21, 2) if n%3==0])
````

**Usando a função `zip`**. Pareando listas obtidas por compreensão.Vamos obter uma lista contendo os pares $(x,\sin{x})$ para alguns valores  $0< x < \pi$. Aqui usamos a função `round` para arredondar para o número mais próximo com 4 casas decimais.

````python
import math

A = [0, 1/4, 1/2, 2.0, round(math.pi,4) ]

B = [round(math.sin(i), 4) for i in A]

AxB = zip(A,B) # pareamento de listas
print(list(AxB))
````

A função `zip` é usada para combinar elementos de duas ou mais listas em pares ordenados. Ela cria um objeto iterável que contém tuplas formadas pelos elementos correspondentes das listas de entrada. Como vimos acima, se tivermos duas listas `A` e `B`, ao usar `zip(A, B)`, obteremos um iterador que retornará tuplas contendo o primeiro elemento de `A` e o primeiro elemento de `B`, o segundo elemento de `A` e o segundo elemento de `B`, e assim por diante.
