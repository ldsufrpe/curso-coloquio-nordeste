## A Estrtágia de Divisão e Conquista

A estratégia de divisão e conquista é uma abordagem algorítmica poderosa utilizada para resolver problemas complexos, decompondo-os em subproblemas menores, resolvendo esses subproblemas de forma recursiva e, em seguida, combinando suas soluções para obter a resposta final.

### Como Funciona

A estratégia de divisão e conquista é composta por três etapas fundamentais:

1. **Divisão:** O primeiro passo é dividir o problema original em subproblemas menores e mais simples. Essa divisão é realizada de maneira recursiva até que os subproblemas atinjam um tamanho que possa ser resolvido de forma direta e trivial.

2. **Conquista:** Após a divisão, a estratégia de conquista é aplicada. Cada subproblema é resolvido de forma recursiva até que se atinja os casos base, onde os problemas são pequenos o suficiente para serem resolvidos diretamente.

3. **Combinação:** Na última etapa, as soluções dos subproblemas são combinadas de maneira eficiente para produzir a resposta final do problema original.

A eficiência da estratégia de divisão e conquista depende da escolha adequada dos casos base, garantindo uma divisão equilibrada dos subproblemas e otimizando a combinação de suas soluções.

## Multiplicação de grandes números inteiros

### Multiplicação padrão

Um algoritmo para multiplicação de números inteiros já é ensinado na escola primária: Para multiplicar dois inteiros positivos $x$ e $y$, você multiplica $x$ por cada dígito de $y$ e organiza os resultados como as linhas de uma tabela, alinhadas sob os dígitos correspondentes de $y$. A soma resulta no produto $x \cdot y$. Aqui está um exemplo

multiplicacion_ejemplo.png

- Um total de 16 multiplicações para um número de 4 dígitos. De fato, para multiplicar dois números, cada um deles com $n$ dígitos serão necessárias $n^2$ multiplicações.


### Algoritmo de Karatsuba

  Dados dois números $x$, $y$ cada um deles consistindo de $n=2^k$ dígitos. Escrevendo ambos na forma
\begin{align*}
    x = a\cdot 10^{n/2} +b \quad \text{e} \quad y = c\cdot 10^{n/2} +d
\end{align*}
temos que
\begin{align*}
    x\cdot y = \left(a\cdot 10^{n/2} +b\right)\left(c\cdot 10^{n/2} +d\right)=ac\cdot 10^n + (ad + bc)\cdot 10^{n/2} + bd.
\end{align*}
Fazendo $u=ac$, $v=bd$ e $w=(a-b)(c-d)$  encontramos $ad + bc = u+v-w$ e então o produto entre $x$ e $y$ é  calculado por
\begin{align*}
    x\cdot y = u\cdot 10^n + (u+v-w)\cdot 10^{n/2} + v.
\end{align*}



### Exemplo 1.
Multiplicação de dois números $x=78$ e $y=21$ com apenas dois dígitos ($n=2$).

Separamos os digitos $x$ e $y$ escrevendo:

$$x=7\cdot 10+ 8$$
$$y = 2\cdot 10 + 1$$

Neste caso, $a=7$, $b=8$, $c=2$, $d=1$. Calculando $u$, $v$ e $w$:

$$u= 7\cdot 2=14$$
$$v = 8\cdot 1=8$$
$$w=(7-8)\cdot(2-1)=-1$$

Então

 $$x\cdot y = 14\cdot 10^2 + (14+8 - (-1))\cdot 10 + 8 = 1638$$



### Exemplo 2.

Multiplicação de dois números $x=4572$ e $y=3169$ com apenas 4 dígitos ($n=4$).

Separamos os digitos $x$ e $y$ escrevendo:

$$x=45\cdot 10^2+ 72$$
$$y = 31\cdot 10^2 + 69$$

Neste caso, $a=45$, $b=72$, $c=31$, $d=69$. Calculando $u, v, w$:

$$ u=\color{red}{45\cdot 31}=1395$$
$$v = \color{red}{72\cdot 69}=4968$$
$$w=\color{red}{(45-72)\cdot(31-69)}=1026$$

$$x\cdot y = 1395\cdot 10^4 + (1395+4968 - 1026)\cdot 10^2 + 4968 = 14488668$$



### Observações

**Recursividade.** Note que poderiamos usar o algoritmo de Karatsuba para calcular cada  produto (destacado acima), já que são numeros de dois dígitos cada. Isso sugere que o algoritmo pode ser usando de forma **recursiva.**

**Número de multiplicações**: Cada multiplicação de números de 2 dígitos requer 3 multiplicações simples (veja Exemplo 1). Portanto, usando recursividade, um número de 4 dígitos requer 3 operações com 2 dígitos. Logo, o total  de operações de multiplicação é igual a $3\cdot 3=9$ .

#### Implementação

```python
# Código aqui
import math

def karatsuba(x, y): # etapa de Caso base
    if x<10 or y < 10:
        return x*y

    #Etapa divisão
    else:

        n = math.floor(math.log10(max(abs(x),abs(y)))) + 1

        m = int(n/2)

        a = x//10**m
        b = x%10**m

        c = y//10**m
        d = y%10**m


        # Recursão

        u = karatsuba(a, c)
        v = karatsuba(b, d)
        w = karatsuba(a-b, c-d)

    #combinação
    return u*10**n + (u + v - w)*10**m + v

```
